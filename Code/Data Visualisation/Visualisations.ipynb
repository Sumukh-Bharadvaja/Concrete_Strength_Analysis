### Domain Analysis

#### We will now check the feautres present in the dataset 

After noting down we can see that all the components are construction materials utilised for the creating Concrete

1. Cement (kg in m^3 mixture) - The amount of cement in kilograms(kg) which is utilised in the meter-cube of the mixture.Cement is a primary ingredient in concrete, providing strength and structural integrity.
2. Blast furnace Slag (kg in a m^3 mixture) - The weight of blast furnace slag in the mixture, measured in kilograms per cubic meter. Blast furnace slag is a byproduct of iron production, used in concrete to improve durability and reduce permeability.
3. Fly Ash (kg in a m^3 mixture) - Amount of fly ash in kilograms per cubic meter. Fly ash, a byproduct of coal combustion in power plants, is used as a supplementary cementitious material. It enhances concrete workability and long-term strength.
4. Water (kg in m^3 mixture) - The amount of water(kg) which is utilised in meter cube of mixture. Water acts as a primary binding agent crucial for chemical reaction(hydration) for cement.
5. Superplasticizer (kg in m^3 mixture) - The weight of the superplasticizer (kg) added to meter cube of the mixture. It is utilised to increase the strength and flowability of mixture
6. Coarse Aggregate (kg in m^3 mixture) - The amount of coarse aggregate (like gravel) in kilograms per cubic meter. Coarse aggregate adds volume, stability, and resistance to wear and erosion in concrete.
7. Fine Aggregate (kg in m^3 mixture) - Measures the fine aggregate (such as sand) in kilograms per cubic meter. Fine aggregate fills spaces between the coarse aggregate and contributes to the concrete's strength and workability.
8. Age (days)- The curing age of the cement. The age is crucial as concrete gains strength over time through the hydration process.

9. Concrete compressive strength(MPa, megapascals)- The compressive strength of the concrete mixture measured in megapascals (MPa). This is a key indicator of the concrete's ability to withstand loads without failing.

#### Since every variable is a continuos feature:
Histogram: A histogram is a bar graph-like representation of data that buckets a range of classes into columns along the horizontal x-axis. The vertical y-axis represents the number count or percentage of occurrences in the data for each column.



## 1. Cement

```python
plt.figure(figsize=(15,8))
sns.histplot(data['Cement'],bins=40)
plt.xlabel('Cement (kg in m^3 mixture)',fontsize=15)
plt.ylabel('Frequency - Count', fontsize=15)
```

# Bivariate Analysis

- It is conducted in order to check for the relation betweeen two variables

### Plots used

1. Line Plot : A Line plot can be defined as a graph that displays data as points or check marks above a number line, showing the frequency of each value.

## 1. Cement vs Compressive Strength

```python
plt.figure(figsize=(20,10))
sns.lineplot(x='Cement',y='CompressiveStrength',data=data,color="Green")
plt.xlabel('Cement (kg in m^3 mixture)',fontsize=15)
plt.ylabel('Compressive Strenth of Concrete (MPa)',fontsize=15)
plt.show()
```

- The plot suggests that there might be a relation between cement and compressive strength , but doesn't confirm the trend of the relation

## 2. Water vs Compressive Strength

```python
plt.figure(figsize=(20,10))
sns.lineplot(x='Water',y='CompressiveStrength',data=data,color="Green")
plt.xlabel('Water (kg in m^3 mixture)',fontsize=15)
plt.ylabel('Compressive Strenth of Concrete (MPa)',fontsize=15)
plt.show()
```

- The graph suggest that there might be a relation between the cement and water but nature of relation is not confirmed

## 3. Coarse Aggregate vs Compressive Strength

```python
plt.figure(figsize=(20,10))
sns.lineplot(x='CoarseAggregate',y='CompressiveStrength',data=data,color="Green")
plt.xlabel('Coarse Aggregate (kg in m^3 mixture)',fontsize=15)
plt.ylabel('Compressive Strenth of Concrete (MPa)',fontsize=15)
plt.show()
```

- There is somewhat similar resembalance between the last graph and this , the nature of relation cannot be explained

## 4. Age vs Compressive Strength

```python
plt.figure(figsize=(20,10))
sns.lineplot(x='Age',y='CompressiveStrength',data=data,color="Green")
plt.xlabel('Age (days)',fontsize=15)
plt.ylabel('Compressive Strenth of Concrete (MPa)',fontsize=15)
plt.show()
```

- From the graph we can conclude that the age does effect the compressive strength. As the number of days increases the compressive strength of mixure also increases.

## Conclusion

- The compressive stregth has a relationn with amount of cement , water , coarse aggregate in the mixture but the type of relation cannot be concluded using the graphs. Further analysis may be done to confirm the type of relationship.
